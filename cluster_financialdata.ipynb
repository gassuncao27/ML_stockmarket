```python
# Target definido
stocks = np.array(dash_fundamentus['Cod'])
alpha = np.array(dash_fundamentus['Perf_pos2t'] - dash_fundamentus['Var_Ibov'])
rank = (alpha - np.min(alpha)) / (np.max(alpha) - np.min(alpha))

Q1 = np.percentile(rank, 25)
Q2 = np.percentile(rank, 50) 
Q3 = np.percentile(rank, 75)
print(f'Q1: {Q1}, Q2: {Q2}, Q3: {Q3}')

def determine_quartile(value):
    if value <= Q1:
        return 1
    elif value <= Q2:
        return 2
    elif value <= Q3:
        return 3
    else:
        return 4
quartile_labels = np.array([determine_quartile(x) for x in rank])

target= np.column_stack((stocks, alpha, rank, quartile_labels))
print(target.shape)
```
